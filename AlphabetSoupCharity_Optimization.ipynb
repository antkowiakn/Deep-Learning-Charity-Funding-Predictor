## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf


#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
cleaned_df=application_df.drop(columns=['EIN', 'NAME'])
cleaned_df
```

```python
# Determine the number of unique values in each column.
for column in cleaned_df:
    print(column, len(cleaned_df[column].unique()))
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_counts= cleaned_df['APPLICATION_TYPE'].value_counts()
application_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#want 10 application_types_to_replace
application_types_to_replace=list(application_counts[application_counts<100].index)
application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    cleaned_df['APPLICATION_TYPE'] = cleaned_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
cleaned_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts= cleaned_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_count= cleaned_df['CLASSIFICATION'].value_counts().loc[lambda x : x>1]
classification_count
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace=list(classification_counts[classification_counts<800].index)
classifications_to_replace
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    cleaned_df['CLASSIFICATION'] = cleaned_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
cleaned_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# referenced supervised machine learning
cleaned_df= pd.get_dummies(cleaned_df, dtype=float)
cleaned_df.head()
```

```python
# Split our preprocessed data into our features and target arrays 
#referencing 21/2/07-detecting diabetes
y = cleaned_df['IS_SUCCESSFUL'].values
X = cleaned_df.drop(columns="IS_SUCCESSFUL").values

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
